##### Breed1, Breed2

```python
breedCheck_Mix = intake['Breed'].unique()
breedCheck_Mix
```

```python
import numpy as np
breed1 = []
breed2 = []

# breedMix.split('/')
for n in  breedCheck_Mix:
  if '/' in n:
    text = n.split('/')
    breed1.append(text[0])
    breed2.append(text[1])
    # print(f'{text[0]}\n {text[1]}')
  else :
    if 'Mix' in n:
      text = [i for i in n.split() if i != "Mix"]
      
      breed1.append(" ".join(text))
      breed2.append(None)
      continue
    breed1.append(n)
    breed2.append(None)
```

##### Decide Final Size
for Size1 and Size2
- if 'value1' & 'value2' : mean
- if 'value1' & 0 : value 1
- if 0 & 0 : 0

```python
breedCheck
```

```python
a = breedCheck['Size 1'] 
b = breedCheck['Size 2']
breedCheck['Final Size'] = 0
for index in range(len(a)):
  if ((a[index] != 0) and (b[index] != 0)):
    finalSize = (a[index]+b[index])/2
  elif ((a[index] == 0) and (b[index] != 0)):
        finalSize = b[index]
  elif ((b[index] == 0) and (a[index] != 0)):
        finalSize = a[index]
  else :
      finalSize = 0
  breedCheck['Final Size'][index] = finalSize
```

```python
MonthYear = intake['MonthYear']

month_intake = []
year_intake = []
for n in MonthYear:
  if ' ' in n:
    text = n.split(' ')
    month_intake.append(text[0])
    year_intake.append(text[1])
```

change Month to int

```python
months_in_year = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
months_int = []
for i in range(1,13):
  months_int.append(i)

monthToInt_dict = dict()
for index in range(len(months_in_year)):
  name = months_in_year[index]
  monthInt = months_int[index]
  monthToInt_dict[name] = monthInt
```

```python
def get_monthInt(monthCol, month_dict):
  if monthCol in month_dict:
    return monthToInt_dict[monthCol]
  else:
    return 0
```
